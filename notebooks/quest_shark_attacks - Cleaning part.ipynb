```python
# # Download the dataset and import it into Python
import xlrd
import pandas as pd
import numpy as np
import re

shark_attacks_df = pd.read_excel('GSAF5.xls')
display(shark_attacks_df.head())

# shark_attacks_df.rename(columns='Species':'species')
display(shark_attacks_df.columns)

country_df = pd.read_excel('country_clean.xls')
display(country_df.head())

df=pd.concat([shark_attacks_df, country_df], axis=1)
display(df.columns)
# Cleaning DataFrame
#Chose the columns we will be working with
df = df[['Date', 'Year', 'Type', 'Country', 'State', 'Location','Activity','Sex', 'Age', 'Injury','Time','Species ','Country_clean']] 
display(df)

```

```python
import poor_bruce as pb
# clean columns titles --> remove spaces & lower cases
df=pb.clean_columns_titles(df)
# drop the null value rows
df=pb.drop_null_values(df)
# format column year
df=pb.format_year(df)
# filtered df
filtered_df=pb.filter_year(df)
filtered_df=pb.format_date(filtered_df)
# filtered_df=pb.create_month(filtered_df)
# filtered_df=pb.create_season(filtered_df)
# filtered_df=pb.fix_sex(filtered_df)
# filtered_df=pb.fix_age(filtered_df)
# final_df=pb.fix_time(filtered_df)
```

```python
def extract_month(date):
    # import re
    # Use regex to extract only the letters
    letters_only = re.findall("[a-zA-Z]+", date)
    # Join the extracted letters into a single string
    month = ''.join(letters_only)
    return month
filtered_df['month']=filtered_df['date'].apply(extract_month)
def check_season(month): 
    q1=['Jan','Feb','Mar']
    q2=['Apr','May','Jun']
    q3=['Jul','Aug','Sep']
    q4=['Oct','Nov','Dec']
    if month in q1:
        season='q1'
    elif month in q2:
        season='q2'
    elif month in q3:
        season='q3'
    elif month in q4:
        season='q4'
    else:
        season="not sure"
    return season 
filtered_df['season']=filtered_df['month'].apply(check_season)


# format column sex
filtered_df[['sex']]=filtered_df[['sex']].fillna("N")
filtered_df['sex']=filtered_df['sex'].str.replace(" ","")

# In column "age" --> Stéphanie
# Define the function to categorize age groups
def set_age_group(x):
    if x == 0:
        return 'Unknown'
    elif x <= 12:
        return 'Child'
    elif x <= 18:
        return 'Teen'
    elif x <= 60:
        return 'Adult'
    elif x > 60:
        return 'Senior'
    else:
        return 'Unknown'

filtered_df['age_0'] = filtered_df['age'].apply(lambda x:int(x) if str(x).strip().replace(".","").isdecimal() else 0)
str(filtered_df['age'][0]).strip().replace(".","").isdecimal()
# filtered_df[filtered_df['age_0']==0]['age'].value_counts().values.sum()
#Seulement 132 valeurs "funky" donc NS =>change it to median 
# Format column age
filtered_df['age'] = pd.to_numeric(filtered_df['age'], errors='coerce')
# Calculate the median age
median_age = filtered_df['age'].median()
print("Median age:", median_age)
# Replace '0' values with the median age
filtered_df['age_0'] = filtered_df['age_0'].replace(0, median_age, inplace=False)
# Apply the age group function to create the 'agegroup' column
filtered_df['age_group'] = filtered_df['age_0'].apply(set_age_group)


# In column "time" --> Kranta
def set_time_group(x):
    if x <=6:
# == [7,8,9,10,11,12]
        return 'Night'
    elif x <=12:
# ==[13,14,15,16,17,18]
        return 'Morning'
    elif x <= 18:
#  [19,20,21,22,23,00]
        return 'Afternoon'
    elif x <=23 :
# [1,2,3,4,5,6]
        return 'Evening'
    else:
        return "Unknown"
    

filtered_df["time"].value_counts().sort_values(ascending=False)
filtered_df["time"] = filtered_df["time"].str.strip()

filtered_df["time"] = filtered_df["time"].str.replace('[^a-zA-Z0-9 ]', '', regex=True)

filtered_df['time']= filtered_df['time'].fillna("Unknown")

filtered_df["time_split"] = filtered_df["time"].str.split('h')

filtered_df["time_split_h"] = filtered_df['time_split'].apply(lambda x:x[0])

filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('1600',"16")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('0500',"5")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('0830',"8")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('0830',"8")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('0830',"8")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Early Morning',"8")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Morning',"8")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Afternoon',"14")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Late afternoon',"14")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Midday',"12")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Early afternoon',"14")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Sunset',"18")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('AM',"10")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('After noon',"14")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Sometime between 06h00  08hoo',"06")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Shortly before 12h00',"11")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Midnight',"01")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('10j30',"10")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Just before noon',"11")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('10jh45',"11")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Just before noon',"11")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Early  morning',"11")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Dusk',"04")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Dawn',"19")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Before 10h00',"09")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Early morning',"6")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Nig',"20")
filtered_df['time_split_h'] = filtered_df['time_split_h'].replace('Evening',"17")

filtered_df['time_split_h_int'] = filtered_df['time_split_h'].apply(lambda x:int(x) if str(x).isdigit() else 26)
filtered_df

filtered_df['time_group'] = filtered_df['time_split_h_int'].apply(set_time_group)
```
